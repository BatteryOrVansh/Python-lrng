```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
month=['jan','feb','mar','apr','may','june']
visitors=[10000,12500,15200,11800,14100,16300]

plt.figure(figsize=(5,4))
plt.xlabel("Month")
plt.ylabel("Website Visitors")
plt.title("Website Traffic")
plt.bar(month, visitors, color='orange', edgecolor = 'black')

plt.show()
```

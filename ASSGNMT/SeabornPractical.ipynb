```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python

data = {'month': ['jan', 'feb','mar','apr','may','jun','july','aug','sep','oct','nov','dec'],
        'temp' : [5.4,7.2,9.4,11.6,14.2,16.7,19.1,18.9,16.7,12.8,9.4,6.7]}
df=pd.DataFrame(data)

plt.figure(figsize=(10,8))
sns.lineplot(x='month',y='temp',data=df)
plt.xlabel('Month')
plt.ylabel('Avg temp (Cel)')
plt.title('Avg Temp in Seattle')
plt.show()
```

```python
data = {'sh' : [4,6,8,2,5,7,3,9,1,4],
        'score': [65,80,92,50,72,88,58,95,42,68]}
df=pd.DataFrame(data)
sns.lmplot(x='sh',y='score',data=df)
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.title('Study Hours vs Exam Score')
plt.show()
```

```python
data = {'Job Title' : ['Manager', 'Engineer', 'Salesperson', 'Manager', 'Engineer', 'Salesperson', 'Manager', 'Engineer', 'Salesperson', 'Manager'],
        'Income' : [85000,72000,60000,90000,68000,55000,100000,75000,48000,88000]}

df = pd.DataFrame(data)
sns.stripplot(x='Job Title', y = 'Income' , data =df)

plt.xlabel('JOB TITLE')
plt.ylabel('INCOME (in USD)')
plt.title('INCOME OF VARIOUS TITLES')
plt.show()
```

```python
sns.swarmplot(x='Job Title', y='Income', data =df)
plt.xlabel('JOB TITLE')
plt.ylabel('INCOME (in USD)')
plt.title('INCOME OF VARIOUS TITLES')
plt.show()
```

```python
data = {'Ratings' : ['Very Satisfied','Satisfied','Neutral','Dissatisfied', 'Very Dissatisfied'],
        'Count' : [250,300,180,80,40]}
df = pd.DataFrame(data)
sns.barplot(x='Ratings', y='Count', data =df )
plt.xlabel('RATINGS')
plt.ylabel('COUNT')
plt.title('CUSTOMER SATISFACTION RATINGS')

plt.show()
```

```python
data= {'Product Category' : ['Electronics', 'Clothing', 'Furniture'],
       'Order Count' : [520,380,210]}
df=pd.DataFrame(data)
sns.countplot(x='Product Category', data =df )
plt.title('Purchase Category')
plt.show()
```

```python
data = {'Fertilizer Type' : ['Organic', 'Organic', 'Organic', 'Synthetic', 'Synthetic', 'Synthetic', 'Organic', 'Organic', 'Synthetic', 'Synthetic'],
        'Plant Height': [20,25,30,18,22,35,28,32,24,38]}
df=pd.DataFrame(data)
sns.violinplot(x='Fertilizer Type', y = 'Plant Height', data = df)
plt.title('HEIGHT DUE TO FERTILIZER')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=[5000,6000,7000,6500,8000]
z=[15,50,90,35,78]
```

```python
bs=[]
for i in z:
    a=i*10
    bs.append(a)
bs
```

```python
plt.figure(figsize=(8,6))
plt.xlabel('population')
plt.ylabel('income')
plt.title('bubble chart')
plt.legend()
plt.scatter(x,y,s=bs,c='orange',alpha=0.8,edgecolors='black')
plt.show()
```

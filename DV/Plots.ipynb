IMAGE CONTOUR

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
img=Image.open(r"C:\Users\Sabreena Khan\Downloads\pandaphoto.jpeg").convert("L")
image_data=np.array(img)
plt.figure(figsize=(10,8))
level=np.linspace(1,10,10)
c=plt.contour(image_data,levels=level,cmap="plasma")
plt.imshow(image_data,cmap="gray",alpha=0.5)
plt.colorbar(c,label="Intensity")
plt.tight_layout()
plt.show()
```

POLAR PLOT

```python
wd=[30,60,90,120,150,180,210,240,270,300,330]
ws=[5,7,8,6,4,3,5,7,9,6,4]
theta=np.radians(wd)
plt.figure(figsize=(10,8))
plt.polar(theta,ws,linestyle="-",color="green",marker="*",markersize=10,markerfacecolor="red")
plt.title("Polar Plot")
plt.xlabel("Wind direction")
plt.ylabel("Wind speed")
plt.show()
```

BOX PLOT

```python
import pandas as pd
data={"col":[78,82,85,88,90,92,93,95,97,100,65,72,150,40]}
df=pd.DataFrame(data)
per25=df["col"].quantile(0.25)
per75=df["col"].quantile(0.75)
iqr=(per75-per25)
ul=per75+(iqr*1.5)
ll=per25-(iqr*1.5)
df=df[df["col"]<=ul]
df=df[df["col"]>=ll]
plt.boxplot(df)
df
```

BUBBLE PLOT

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=[5000,6000,7000,6000,9000]
z=[15,30,82,68,95]
bs=[]
for i in z:
    a=i*50
    bs.append(a)
plt.figure(figsize=(8,6))
plt.scatter(x,y,s=bs,c="yellow",alpha=0.8,edgecolor="black")
plt.xlabel("population")
plt.ylabel("income")
plt.title("Bubble plot")
plt.legend()
plt.show()
```

WAFFLE CHART

```python
import matplotlib.pyplot as plt
import pandas as pd
from pywaffle import Waffle
data={"Phone":["Mi","Apple","Samsung","Nokia","Realme"],
      "Stocks":[44,12,8,5,3]}
df=pd.DataFrame(data)
fig=plt.figure(figsize=(8,9),FigureClass=Waffle,rows=5,values=df.Stocks,label=list(df.Phone))
plt.title("Waffle Chart")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

stock = [100, 30, 75, 50, 150, 125, 160, 90]
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007]
trading_volume = [14, 16, 19, 15, 25, 20, 17, 22]

fig, ax1 = plt.subplots()

# Plotting stock prices on the primary y-axis
ax1.fill_between(years, stock, color='blue', alpha=0.5, label="Stock Prices")
ax1.set_xlabel('Years')
ax1.set_ylabel('Stock Prices', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Creating a secondary y-axis for trading volume
ax2 = ax1.twinx()
ax2.fill_between(years, trading_volume, color='red', alpha=0.5, label='Trading Volume')
ax2.set_ylabel('Trading Volume', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Add legend
plt.legend()

plt.title('Stock Prices and Trading Volume Over Years')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for six countries
countries = ['Country 1', 'Country 2', 'Country 3', 'Country 4', 'Country 5', 'Country 6']
population = [10, 45, 33, 150, 7, 300]  # Population in millions
area = [0.2, 0.5, 1.1, 3.2, 1, 2.5]  # Area in million square kilometers
gdp = [0.8, 1, 1.2, 5, 0.6, 3.3]  # GDP in trillion USD

# Calculate population density (population per square kilometer)
population_density = [pop / a for pop, a in zip(population, area)]

# Create bubble chart
plt.figure(figsize=(10, 6))
bubble_sizes = [pop * 10 for pop in population]  # Adjust bubble sizes for better visualization
plt.scatter(gdp, population, s=bubble_sizes, c=population_density, cmap='viridis', alpha=0.6, edgecolors='w', linewidth=1)

# Add labels and title
plt.xlabel('GDP (Trillions of USD)')
plt.ylabel('Population (Millions)')
plt.title('Bubble Chart: GDP, Population, and Population Density')

# Add colorbar for population density
cbar = plt.colorbar()
cbar.set_label('Population Density (per sq km)')

# Add country labels
for i, country in enumerate(countries):
    plt.text(gdp[i], population[i], country)

plt.grid(True)
plt.show()
```

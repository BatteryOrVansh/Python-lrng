```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python


ic=np.random.rand(100)
fc=np.random.randint(100*0.1)

sns.distplot(ic)
```

```python
#line plot

temp=np.random.randint(25,45,10)
days=np.arange(1,11)

sns.lineplot(x=days,y=temp)

# plt.plot(temp)
```

```python
#lm plots


data = {'x':[1,2,3,4,5],
        'y':[4,5,6,8,2]}

df=pd.DataFrame(data)

sns.lmplot(x ='x',y='y',data=df)
```

```python
# categorical plots
# strip plot


data = {'category' : ['A','A','B','B','C'],
        'value' : [2,2,1,4,7]}
df=pd.DataFrame(data)
sns.stripplot(x='category',y='value',data=df)
```

```python
# swarm plots

data = {'category' : ['A','A','B','B','C'],
        'value' : [2,2,1,4,7]}
df=pd.DataFrame(data)
sns.swarmplot(x='category',y='value',data=df)
```

```python
# count plot

data = {'category' : ['A','A','B','B','C'],
        'value' : [2,5,1,4,7]}
df=pd.DataFrame(data)

sns.countplot(x='category',data=df)
```

```python
# violin plot


data = {'category' : ['A','A','B','B','C'],
        'value' : [2,5,1,4,7]}
df=pd.DataFrame(data)


sns.violinplot(x='category',y='value',data=df)
```

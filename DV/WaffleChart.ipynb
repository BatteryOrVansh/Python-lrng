```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle
```

```python
data={'phone':['samsung','apple','xiaomi','oneplus','nothing'],
      'stock':[25,30,20,15,10]}
df=pd.DataFrame(data)
fig=plt.figure(figsize=(10,10),FigureClass=Waffle,rows=5,values=df['stock'],labels=list(df['phone']))
plt.title('Waffle Chart')

plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(8,6)) #defines the figure size 
ax = fig.add_subplot(projection='3d') #converting the axes into 3d
```

```python


# to show the dashboard of multiple charts at one place

# fig and ax defined in single line only... there's no projection 3d
fig, ax = plt.subplots(3,2, figsize=(10,5))

#data points
x=[1,2,3,4]
y=[3,4,2,3]

#plotting line plot
ax[0,0].plot(x,y) #ax[0,0] slices the charts and selects the chart from 0th row and 0th column

#plotting the scatter plot
ax[1,0].scatter(x,y) # selects chart from 1st row and 0th column\

plt.show()
```

```python
## Scatter Plot in 3d

x=[1,2,3,4,5]
y=[4,6,5,8,2]
z=[7,3,1,5,2]

fig=plt.figure(figsize=(10,8))

ax=fig.add_subplot(111,projection='3d')

ax.plot(x,y,z)
plt.show()
```
